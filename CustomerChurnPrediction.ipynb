```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Load Dataset
df = pd.read_csv(r"C:\Users\Awadhesh Kumaar\Downloads\Churn_Modelling.csv")
```

```python
# show the first 5 data
df.head()
```

```python
df.isnull().sum()
```

```python
# Remove the unnecesary columns 
df = df.drop(["RowNumber","CustomerId","Surname"],axis =1)
```

```python
# Encode Categorical Variable
le_gender = LabelEncoder() # converting text data into numbers
df["Gender"] = le_gender.fit_transform(df["Gender"])
df = pd.get_dummies(df,columns=["Geography"],drop_first=True)# One hot encoding
```

```python
# Features and Target
x = df.drop("Exited",axis = 1)# Removing only exited data from the table
# if here mention axis = 0 they remove the rows data
y = df["Exited"]
```

```python
print(x)
```

```python
print(y)
```

```python
# Train - Test - Split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Feature Scalling
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Build ANN Model
model = Sequential()

model.add(Dense(16,activation='relu',input_dim = x_train.shape[1]))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
# Compile model
model.compile(
    optimizer = 'adam', 
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
# train model
history = model.fit(
    x_train, y_train, epochs = 50, batch_size = 32, validation_split=0.2)
```

```python
# User Input

creditscore = int(input("Enter Credit Score(350 - 850): "))
gender = input("Gender (Male/Female): ").capitalize()
age = int(input("Enter Age(18 - 92): "))
tenure = int(input("Enter Tenure(0-10): "))
balance = float(input("Enter Balance(0-250898): "))
numofproducts = int(input("Enter Number of Products(1-4): "))
hascrcard = int(input("Has Credit Card (1/0): "))
isactivemember = int(input("Is Active Member (1/0): "))
estimatedsalary = float(input("Enter Estimated Salary(11-199992): "))
geography = input("Geography (France/Germany/Spain): ").capitalize()
```

```python
# Gender Encoding

if gender == "Male":
    gender = 1
else:
    gender = 0

# Geography Encoding

geography_germany = 0
geography_spain = 0

if geography == "Germany":
    geography_germany = 1
elif geography == "Spain":
    geography_spain = 1
```

```python
# user data
user_data = [[
    creditscore,
    gender,
    age,
    tenure,
    balance,
    numofproducts,
    hascrcard,
    isactivemember,
    estimatedsalary,
    geography_germany,
    geography_spain
]]
```

```python
# Scale data
user_data = scaler.transform(user_data)
```

```python
# Evaluate Model
loss, accuracy = model.evaluate(x_test, y_test)
print("Accuracy:",accuracy)
```

```python
# Predictions

Predictions = model.predict(user_data)
Predictions = (Predictions > 0.5)
print(Predictions)
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix, classification_report
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```
